```python
import sqlite3
## connect to sqlite
conn = sqlite3.connect('customer.db')
```

```python
# creating a cursor object
cursor = conn.cursor()

table_info = """
CREATE TABLE customer (
    customer_id VARCHAR(20),
    platform VARCHAR(20),
    segment VARCHAR(20),
    SKU VARCHAR(50),
    hour INT,
    date DATE,
    ROAS FLOAT,
    CTR FLOAT,
    CPC FLOAT,
    CPM FLOAT,
    CPA FLOAT,
    impressions INT,
    clicks FLOAT,
    conversions FLOAT,
    spend FLOAT,
    revenue FLOAT,
    ad_name VARCHAR(100),
    ad_type VARCHAR(20),
    campaign_id VARCHAR(50)
);
"""
# executing the cursor object
cursor.execute(table_info)
```

```python
import csv

# Open the CSV file and insert records
with open('customerHourly.csv', 'r', newline='') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header row
    cursor.executemany("""
        INSERT INTO customer (
            customer_id, platform, segment, SKU, hour, date, ROAS, CTR, CPC, CPM, CPA,
            impressions, clicks, conversions, spend, revenue, ad_name, ad_type, campaign_id
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
    """, reader)

# Commit and close
conn.commit()
conn.close()
```
